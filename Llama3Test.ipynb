**Instaling the Dependencies**

```python
import json
import torch
from transformers import(AutoTokenizer,
                         AutoModelForCausalLM,
                         BitsAndBytesConfig,
                         pipeline)
```

**HF Account Configuration**

```python
#config_data = json.load(open("config.json"))
#HF_TOKEN = config_data["HF_TOKEN"]
from google.colab import userdata
HF_TOKEN = userdata.get("HF_TOKEN")
print(HF_TOKEN)
model_name = "meta-llama/Meta-Llama-3.1-8B"
```

**Quantisation Configuration** model jest 8B -> 32GB, redukcja do 4B -> 16GB

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

**Loading Tokenizer and LLM**

```python
tokenizer = AutoTokenizer.from_pretrained(model_name, token=HF_TOKEN)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    token=HF_TOKEN
)
```

**Prompt into LLM**

```python
text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)
```

```python
prompt = "Dlaczego Podlasie jest super?"
#test
response = text_generator(prompt)
response
```

```python
def get_response(prompt):
  sequences = text_generator(prompt)
  gent_text = sequences[0]["generated_text"]
  return gent_text
```

```python
#so it is !!! :-)
llama3_response = get_response(prompt)
print(llama3_response[len(prompt):])
```
